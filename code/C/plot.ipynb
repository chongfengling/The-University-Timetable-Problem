```python
def tabcol_res_vis(filename):
    num_ver = []
    num_col = []
    num_ite = []
    num_tim = []
    with open(filename,'r') as f:
        res = f.readlines()
        for i in range(3, len(res), 4):
            tmp = res[i][1:-2].split(', ')
            num_ver.append(int(tmp[1]))
            num_col.append(int(tmp[2]))
            num_ite.append(int(tmp[3]))
            num_tim.append(float(tmp[4]))
    return num_ver, num_col, num_ite, num_tim

v, c, i, t = tabcol_res_vis('result/TABUCOL_res_fyp1.txt')
```

```python
from mpl_toolkits.axisartist.parasite_axes import HostAxes, ParasiteAxes
import matplotlib.pyplot as plt


fig = plt.figure()

host = fig.add_axes([0.15, 0.1, 0.65, 0.8], axes_class=HostAxes)
par1 = ParasiteAxes(host, sharex=host)
par2 = ParasiteAxes(host, sharex=host)
host.parasites.append(par1)
host.parasites.append(par2)

host.axis["right"].set_visible(False)

par1.axis["right"].set_visible(True)
par1.axis["right"].major_ticklabels.set_visible(True)
par1.axis["right"].label.set_visible(True)

# par2.axis["right2"] = par2.new_fixed_axis(loc="right", offset=(60, 0))

p1, = host.plot(v,c, label="Number of Vertices")
# p3, = par2.plot(v,i, label="vi")
p2, = par1.plot(v,t, label="Time")

# host.set_xlim(t)
# host.set_ylim()
# par1.set_ylim(0, 4)
# par2.set_ylim(1, 65)

host.set_xlabel("Number of Vertices")
host.set_ylabel("Number of Colors")
# par2.set_ylabel("Number of iterations")
par1.set_ylabel("Time")

host.legend()

host.axis["left"].label.set_color(p1.get_color())
par1.axis["right"].label.set_color(p2.get_color())
# par2.axis["right2"].label.set_color(p3.get_color())
plt.title('Random Graph Coloring by TABCOL Algorithm')
plt.show()
```
